```python
import os
import chromadb
from llama_index.vector_stores.chroma import ChromaVectorStore
from llama_index.core import StorageContext, VectorStoreIndex
from llama_index.llms.openai import OpenAI

os.environ["OPENAI_API_KEY"] = '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'

# Path to the persisted ChromaDB directory
persist_directory = "/content/drive/My Drive/Colab Notebooks/data/chromadb"

# Reinitialize ChromaDB Persistent Client
db = chromadb.PersistentClient(path=persist_directory)

# Load the existing collection
collection_name = "100pitches-embeddings"
chroma_collection = db.get_or_create_collection(collection_name)

# Create ChromaVectorStore and StorageContext
vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
storage_context = StorageContext.from_defaults(vector_store=vector_store)

# load your index from stored vectors
index = VectorStoreIndex.from_vector_store(
    vector_store, storage_context=storage_context
)
```

```python
# Convert the index to a retriever
retriever = index.as_retriever(similarity_top_k=100)

# Define your query
query_text = 'Who are the founders or who is the team for each of the startups?'

# Perform the retrieval
response = retriever.retrieve(query_text)

# Print the response
print("Query Response:")
print(response)

# Print a readable summary of the retrieved nodes
print("Query Response:\n")
for idx, node_with_score in enumerate(response, start=1):
    node = node_with_score.node  # Access the node
    print(f"Result {idx}:")
    print(f"Node ID: {node.id_}")
    print(f"Startup: {node.metadata.get('startup', 'Unknown')}")  # Extract metadata if available
    print(f"Text: {node.text}...")  # Print the first 300 characters of the text
    print(f"Score: {node_with_score.score}")
    print("-" * 40)
```

```python
from llama_index.core.response_synthesizers import ResponseMode
from llama_index.core import get_response_synthesizer
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.llms.openai import OpenAI

llm = OpenAI(model="gpt-4")

# Initialize the response synthesizer
response_synthesizer = get_response_synthesizer(response_mode=ResponseMode.COMPACT)

# Create a query engine with the synthesizer and set top_k for the retriever
query_engine = index.as_query_engine(
    response_synthesizer=response_synthesizer,
    similarity_top_k=500,
    llm=llm
)
# Define your query
# query_text = 'Who are the founders or who is the team for all of the startups (Rencore, Silana, Fyta, Electryone, Lyntics)?'
query_text = 'In your context there are informations on 100 independent startups, each of them having multiple chunks in the context. Make a numbered list with all their names. List each startup only once in alphabethical order.'
# query_text = 'You are an investment manager at a venture capital firm. You see the infromations provided for each of the startups, please choose the startups that are worth having a first call with to better understand their business. Explain based on what reasoning you are making this choice.'

# Perform the query
# response = query_engine.query(query_text)
response = query_engine.query(ideal_prompt)

# Print the synthesized response
print("Synthesized Response:")
print(response)
```
